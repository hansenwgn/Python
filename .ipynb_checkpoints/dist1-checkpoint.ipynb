```python
import pandas as pd
import geopandas as gpd
from geopy.geocoders import GoogleV3
from matplotlib_scalebar.scalebar import ScaleBar
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.anchored_artists import AnchoredSizeBar
import matplotlib.font_manager as fm
from shapely.geometry import Point
import plotly.graph_objects as go
from shapely.geometry import LineString
from geopandas.tools import sjoin
from pyproj import Transformer
import matplotlib.colors as mcolors
import matplotlib.cm as cm
import numpy as np
from matplotlib.colors import BoundaryNorm
from matplotlib.colors import LinearSegmentedColormap, Normalize
import matplotlib.patches as mpatches
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv('ttm.csv')
```

```python
df.head(5)
```

```python
df.info()
```

```python
# Group by 'msoa' and find the minimum travel time for each MSOA
min_travel_time_per_msoa = df.groupby('from_id')['travel_time_p50'].min().reset_index()
```

```python
min_travel_time_per_msoa.head(5)
```

```python
min_travel_time_per_msoa.to_csv('mintravel.csv')
```

```python
min_travel_time_per_msoa.info()
```

```python
# Load the geographical data for London MSOA boundaries
gdf = gpd.read_file('ESRI/MSOA_2011_London_gen_MHW.shp')
```

```python
# london borough 
london_borough = gpd.read_file('ESRI/London_Borough_Excluding_MHW.shp')
```

```python
gdf.plot()
```

```python
gdf.head(5)
```

```python
gdf.info()
```

```python
# Merge the data with the MSOA shapefile using a left join
merged_gdf = gdf.merge(df, left_on="MSOA11CD", right_on="from_id", how="left")
```

```python
merged_gdf.info()
```

```python
merged_gdf.head(3)
```

```python
merge1 = gdf.merge(min_travel_time_per_msoa, left_on="MSOA11CD", right_on="from_id", how="left")
```

```python
merge1.head(3)
```

```python
merge1.to_csv('merge1.csv')
```

```python
merge1.info()
```

```python
# read data musical venues
music = pd.read_csv('Music_venues_all.csv')
```

```python
music.head(3)
```

```python
music.info()
```

```python
music_geom = gpd.GeoDataFrame(
    music,
    geometry=gpd.points_from_xy(
        music.longitude, music.latitude
    ),
)
music_geom.head(3)
```

```python
music_geom.crs = {"init": "epsg:4326"}
music_geom = music_geom.to_crs(epsg=27700)
```

```python
#Plotting
fig, ax = plt.subplots(figsize=(20, 20))

# MSOA
gdf.plot(ax=ax, color="gainsboro", edgecolor='white', linewidth=0.5)

# variable
cax = merge1.plot(column="travel_time_p50", ax=ax, cmap='magma')

# normalise
norm = mcolors.Normalize(vmin=merge1["travel_time_p50"].min(), vmax=merge1["travel_time_p50"].max())

# ScalarMappable
mappable = cm.ScalarMappable(norm=norm, cmap='magma')

# London boroughs 
london_borough.plot(ax=ax, color="none", edgecolor='white', linewidth=1)

# music venues
music_geom.plot(ax=ax, markersize = 6, color="red") 

# scale bar
scalebar = ScaleBar(dx=1.0, location='lower right')
ax.add_artist(scalebar)
x, y, arrow_length = 0.05, 0.95, 0.1
ax.annotate('N', xy=(x, y), xytext=(x, y-arrow_length),
            arrowprops=dict(facecolor='black', arrowstyle='wedge'),
            ha='center', va='center', fontsize=20,
            xycoords=ax.transAxes)

ax.axis('off')

# fig
fig.colorbar(mappable, ax=ax, shrink=0.5).set_label('Travel time (mins)', size=12)

plt.title('The minimum travel time to musical venues within 45mins')
plt.show()
```
